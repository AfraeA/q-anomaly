```python
import re
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    'font.size' : 16,
    "figure.dpi": 200,
})
from itertools import product
import os
```

```python
def retrieve_result_per_size_dfs(n_pc=2):
    methods ={"cRBF":"RBF", "qIT":"IT", "qRM_mitig":"Mitigated RM", "qRM_nomitig":"Unmitigated RM", "qVS_max":"VS Max", "qVS_avg": "VS Average"}
    columns = ["size", "avgPrecision", "precision", "recall", "f1_score", "auroc", "train_time", "test_time"]
    df = pd.DataFrame(columns=columns)
    for method in methods.keys():
        file_names = os.listdir(f"./Results/{method}")
        for file in file_names:
            method_df = pd.read_csv(f"./Results/{method}/{file}")
            method_df = method_df[method_df['n_pc']==n_pc]
            method_df['size'] = int(re.findall('\d+', file.split('_')[1])[0])
            method_df['kmethod'] = method
            df = pd.concat([df, method_df], axis=0, ignore_index=True)
    df = df.sort_values('size')
    return {methods[method]: df[df['kmethod']==method] for method in methods}
```

```python
def plot_performance_by_size(smoothing=True, errorbar=True, alpha=0.2, n_pc=2, qIT_shots=None, \
                qRM_shots=None, qRM_settings=None, qVS_subsamples=None):
    result_dfs = retrieve_result_per_size_dfs(n_pc)
    metrics = np.array([['avgPrecision', 'f1_score'], ['precision', 'recall']])
    metrics_nicelywritten={'avgPrecision':'Average precision', 
                           'f1_score':'F1 score', 
                           'precision':'Precision', 
                           'recall':'Recall',
                           'auroc':'AUROC',
                           'train_time':'Training Time [s]',
                           'test_time':'Testing Time [s]'
                          }
    z_orders = {"RBF":5,"IT":-1, "Mitigated RM":2, "Unmitigated RM":-3, "VS Max":-2, "VS Average":3}
    fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12,12))
    fig.tight_layout(pad=2.2)
    for kmethod, df in result_dfs.items():
        g_df = df.groupby('size')
        for i,j in product(range(metrics.shape[0]), range(metrics.shape[1])):
            if smoothing:
                metric_smoothed_mean = g_df.mean()[metrics[i,j]].ewm(alpha=alpha).mean()
            else:
                metric_smoothed_mean = g_df.mean()[metrics[i,j]]
            metric_smoothed_mean.plot(label=kmethod, linestyle='solid', marker='.', ax=axes[i,j], grid=True)
            if errorbar:
                if smoothing:
                    metric_smoothed_max = g_df.max()[metrics[i,j]].ewm(alpha=alpha).mean()
                    metric_smoothed_min = g_df.min()[metrics[i,j]].ewm(alpha=alpha).mean()
                else:
                    metric_smoothed_max = g_df.max()[metrics[i,j]]
                    metric_smoothed_min = g_df.min()[metrics[i,j]]    
                axes[i,j].fill_between(g_df['size'].mean().values, metric_smoothed_min, metric_smoothed_max, alpha=0.15, zorder=z_orders[kmethod])
                axes[i,j].set_xticks(list(range(250, 1750, 250)))
                axes[i,j].set(xlabel='Training data size',ylabel=metrics_nicelywritten[metrics[i,j]])
                if i!=0 or j!=0:
                    axes[i,j].set(ylim=(0,1))
    handles, labels = axes[0,0].get_legend_handles_labels()
    fig.legend(handles, labels, fontsize=16,loc='upper center', ncol=6, bbox_to_anchor=(0.5,1.05))
    #plotsFolderName = f'./Plots/performance_by_size/'
```

```python
plot_performance_by_size()
```

```python
def plot_time_by_size(smoothing=True, errorbar=True, alpha=0.2, n_pc=2, qIT_shots=None, \
                qRM_shots=None, qRM_settings=None, qVS_subsamples=None):
    result_dfs = retrieve_result_per_size_dfs(n_pc)
    metrics = np.array(['train_time','test_time'])
    metrics_nicelywritten={'avgPrecision':'Average precision', 
                           'f1_score':'F1 score', 
                           'precision':'Precision', 
                           'recall':'Recall',
                           'auroc':'AUROC',
                           'train_time':'Training Time [s]',
                           'test_time':'Testing Time [s]'}
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12,6))
    fig.tight_layout(pad=2.2)
    for kmethod, df in result_dfs.items():
        g_df = df.groupby('size')
        for i, metric in enumerate(metrics):
            if smoothing:
                metric_smoothed_mean = g_df.mean()[metric].ewm(alpha=alpha).mean()
            else:
                metric_smoothed_mean = g_df.mean()[metric]
            metric_smoothed_mean.plot(label=kmethod, linestyle='solid', marker='.', ax=axes[i], grid=True)
            if errorbar:
                if smoothing:
                    metric_smoothed_max = g_df.max()[metric].ewm(alpha=alpha).mean()
                    metric_smoothed_min = g_df.min()[metric].ewm(alpha=alpha).mean()
                else:
                    metric_smoothed_max = g_df.max()[metric]
                    metric_smoothed_min = g_df.min()[metric]    
                axes[i].fill_between(g_df['size'].mean().values, metric_smoothed_min, metric_smoothed_max, alpha=0.1)
                axes[i].set_xticks(list(range(250, 1750, 250)))
                axes[i].set(xlabel='Training data size',ylabel=metrics_nicelywritten[metric])
    handles, labels = axes[0].get_legend_handles_labels()
    fig.legend(handles, labels, fontsize=12,loc='upper center', ncol=6, bbox_to_anchor=(0.5, 1.05))
```

```python
plot_time_by_size()
```

```python
methods = {'cRBF':'dsize_{size}.csv', \
           'qIT':'dsize_{size}_n_shots_1000.csv', \
           'qRM_mitig':'dsize_{size}_n_shots_9000_n_settings_30.csv', \
           'qRM_nomitig':'dsize_{size}_n_shots_9000_n_settings_30.csv', \
           'qVS_avg':'dsize_{size}_n_subsamples_{n_subsamples}_maxsize_100.csv', \
           'qVS_max':'dsize_{size}_n_subsamples_{n_subsamples}_maxsize_100.csv'}
sizes = list(range(250, 1750, 250))
for kmethod, name_template in methods.items():
    for size in sizes:
        file_name = name_template.format(size=size, n_subsamples=int(size/100) if kmethod.startswith('qVS') else '')
        if kmethod == 'qVS_avg':
            uncorr_df = pd.read_csv(f"./Results_uncorr/qVS_max/{file_name}")
        else:
            uncorr_df = pd.read_csv(f"./Results_uncorr/{kmethod}/{file_name}")
        uncorr_df = uncorr_df[['seed','n_pc', 'train_time', 'test_time']]
        corr_df = pd.read_csv(f"./Results/{kmethod}/{file_name}")
        corr_df = corr_df.drop(['train_time', 'test_time'], axis=1)
        merge_df = pd.merge(left=corr_df, right=uncorr_df, on=['seed','n_pc'])
        resultsFolderName = f"./Results_final/{kmethod}/"
        if not os.path.exists(resultsFolderName):
            os.makedirs(resultsFolderName)
        if not os.path.exists(f"./Results_final/{kmethod}/{file_name}"):
                merge_df.to_csv(f"./Results_final/{kmethod}/{file_name}", index=False)
```
